## What Are Stacks?

```python
cardStack =  []
cardStack.append('5 of diamonds')
print(', '.join(cardStack))
```

```python
cardStack.append('3 of clubs')
print(', '.join(cardStack))
```

```python
cardStack.append('ace of hearts')
print(', '.join(cardStack))
```

```python
cardStack.pop() 
print(', '.join(cardStack))
```
