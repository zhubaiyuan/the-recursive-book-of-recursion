## Code before and after the recursive call

```python
def countDownAndUp(number):
    print(number)
    # BASE CASE
    if number == 0:
        print('Reached the base case.')
        return
    # RECURSIVE CASE
    else:
        countDownAndUp(number - 1)
        print(number, 'returning')
        return
```

```python
countDownAndUp(3)
```
