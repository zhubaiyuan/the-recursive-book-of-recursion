## Converting an Iterative Algorithm into a Recursive Algorithm

```python
print('Code in a loop:')
i = 0
while i < 5:
    print(i, 'Hello, world!')
    i = i + 1
```

```python
print('Code in a function:')


def hello(i=0):
    print(i, 'Hello, world!')
    i = i + 1
    # RECURSIVE CASE
    if i < 5:
        hello(i)
    # BASE CASE
    else:
        return


hello()
```
