## Reversing a String

```python
def rev(theString):
    # BASE CASE
    if len(theString) == 0 or len(theString) == 1:
        return theString
    # RECURSIVE CASE
    else:
        head = theString[0]
        tail = theString[1:]
        return rev(tail) + head
```

```python
print(rev('abcdef'))
```

```python
print(rev('Hello, world!'))
```

```python
print(rev(''))
```

```python
print(rev('X'))
```
