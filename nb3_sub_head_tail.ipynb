## Summing Numbers in an Array

```python
def sum(numbers):
    # BASE CASE
    if len(numbers) == 0:
        return 0
    # RECURSIVE CASE
    else:
        head = numbers[0]
        tail = numbers[1:]
        return head + sum(tail)
```

```python
nums = [1, 2, 3, 4, 5]
print('The sum of', nums, 'is', sum(nums))
```

```python
nums = [5, 2, 4, 8]
print('The sum of', nums, 'is', sum(nums))
```
