```python
import turtle

# Increase the first argument to speed up the drawing.
turtle.tracer(10, 0)
turtle.setworldcoordinates(0, 0, 700, 700)
turtle.hideturtle()
# Try changing the line length by a little.
LINE_LENGTH = 5
# Try changing the turning angle by a few degrees.
ANGLE = 90
# Try changing the recursive level by a little.
LEVELS = 6
DRAW_SOLID = False
# Uncomment this line to draw the curve at an angle.
# turtle.setheading(20)
```

```python
def hilbertCurveQuadrant(level, angle):
    # BASE CASE
    if level == 0:
        return
    # RECURSIVE CASE
    else:
        turtle.right(angle)
        hilbertCurveQuadrant(level - 1, -angle)
        turtle.forward(LINE_LENGTH)
        turtle.left(angle)
        hilbertCurveQuadrant(level - 1, angle)
        turtle.forward(LINE_LENGTH)
        hilbertCurveQuadrant(level - 1, angle)
        turtle.left(angle)
        turtle.forward(LINE_LENGTH)
        hilbertCurveQuadrant(level - 1, -angle)
        turtle.right(angle)
        return
```

```python
def hilbertCurve(startingPosition):
    # Move to starting position.
    turtle.penup()
    turtle.goto(startingPosition)
    turtle.pendown()
    if DRAW_SOLID:
        turtle.begin_fill()
    # Draw lower left quadrant.
    hilbertCurveQuadrant(LEVELS, ANGLE)
    turtle.forward(LINE_LENGTH)
    # Draw lower right quadrant.
    hilbertCurveQuadrant(LEVELS, ANGLE)
    turtle.left(ANGLE)
    turtle.forward(LINE_LENGTH)
    turtle.left(ANGLE)
    # Draw upper right quadrant.
    hilbertCurveQuadrant(LEVELS, ANGLE)
    turtle.forward(LINE_LENGTH)
    # Draw upper left quadrant.
    hilbertCurveQuadrant(LEVELS, ANGLE)
    turtle.left(ANGLE)
    turtle.forward(LINE_LENGTH)
    turtle.left(ANGLE)
    if DRAW_SOLID:
        turtle.end_fill()
```

```python
hilbertCurve((30, 350))
turtle.exitonclick()
```
